```
import requests
import pandas as pd
import csv
dictionary={}
dictionary["number"]=[]
dictionary["area"]=[]
dictionary["floor"]=[]
dictionary["price"]=[]
amount=0
dictionary["city"]=[]
dictionary["region"]=[]
page=2
f=csv.writer(open('apartments.csv','w',newline=''))
f.writerow(['amount','area','floor','price','city','region'])
while amount<10000:
    results = soup.find_all('a', attrs={'class':'a-card__title'})
    prices = soup.find_all('span', attrs={'class':'a-card__price-text'})
    index=0
    cities=soup.find_all('div',class_='a-card__subtitle')
    for i in results:
        try:
          area=str(i)[str(i).find(",")+2:str(i).find("м²")-1]
        #dictionary["area"].append(area)
        
          floor=str(i)[str(i).find("м²,")+4:str(i).find("этаж")-1]
        
          if(floor.find('class="a-card__title link"')!=-1):
              floor=''
          price=str(prices[index])[str(prices[index]).find(">")+1:str(prices[index]).find("<",1)-1].strip().replace(' ','')
        #dictionary["price"].append(str(price))
        #print(dictionary["price"])
          region=str(cities[index])[str(cities[index]).find(">")+2:str(cities[index]).find("<",2)]
          city=region.split(',')[0].strip()
        #print(region)
        #dictionary["city"].append(str(city))
          region1=region.split(',')[1]
          #dictionary["region"].append(str(region1))
        
        #print(dictionary["city"])
        #print(dictionary["region"])
          index+=1
          amount+=1
          f.writerow([amount,float(area),str(floor),str(price),str(city),str(region1)])
        #dictionary["number"].append(amount)
        except:
          continue
        if amount==10000:
            break
    r = requests.get('https://m.krisha.kz/arenda/kvartiry/?das[rent.period]=2&page='+str(page))
    soup = BeautifulSoup(r.text, 'html.parser')
    page+=1
    print(page,"/",amount)
    
    
records = []

```
